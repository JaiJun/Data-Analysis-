# 練習
## 美國嬰兒姓名統計資料
檔案大小約7mb，如果同學覺得檔案太大不方便匯入，可以先慢慢下載下來後執行，或刪除檔案中的部分資料再匯入練習。<br>
資料下載網址：https://raw.githubusercontent.com/Code-Gym/python-dataset/master/US_Baby_Names_Lite.csv

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Code-Gym/python-dataset/master/US_Baby_Names_Lite.csv')
```

### 請找出名字是Emma的資料

```python
Emma_df = df[df['Name'] == 'Emma']
display(Emma_df)
```

### 請找出名字是An開頭，並且是女性的資料
請用兩個變數建立篩選條件，分別是mask1和mask2，然後再使用符號&，合併條件過濾結果

```python
mask1 = df['Name'].str.startswith('An')
mask2 = df['Gender'] == "F"
res = df[mask1 & mask2]
display(res)
```

### 請找出2004年，最多美國女生取的「菜市場」名

```python
year_filter = df['Year'] == 2004 #過濾年份
gender_filter = df["Gender"] == 'F' #過濾女性
df = df[year_filter & gender_filter] #條件式寫進Pandas
res = df.groupby('Name').Count.sum().sort_values(ascending=False).head(1) #排序取出第一筆資料
print(res)
```
