```python
import pandas as pd
```

## 讀取資料

```python
"""
    age: 年紀
    gender: 性別
    occupation: 職業
    zip_code: 郵政編碼
"""   
就業資料_df = pd.read_excel("就業資料.xlsx",index_col='user_id')
display(就業資料_df.info()) # 整份df 類型 描述出來
display(就業資料_df.describe()) #整份df 有關數值 基本參數都會算出來
```

## 取得每一個職業的年齡最大值,平均值,最小值

```python
職業_平均值 = 就業資料_df.groupby('occupation').age.mean()
職業_最大值 = 就業資料_df.groupby('occupation').age.max()
職業_最小值 = 就業資料_df.groupby('occupation').age.min()
# 職業_平均值
# 職業_最大值
# 職業_最小值
```

## 取得每一個職業的年齡平均值

```python
就業資料_df.groupby('occupation').age.agg('mean')
```

## 取得每一個職業的年齡最小值、最大值、平均值和數量

```python
就業資料_df.groupby('occupation').age.agg(['min','max','mean','count'])
# 就業資料_df.groupby('occupation').agg({'age':['min','max','mean','count']})
```
